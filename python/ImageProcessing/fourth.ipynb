```python
import cv2,numpy as np
```

Events


```python
image = np.zeros((512,512,3)) 
```

```python
def draw(event,x,y,flags,params):
    if event==0:
        print("Mouse Hover")
    elif event==1:
        print("Left click press")
    elif event==7:
        print("Left double click press")
    elif event==4:
        print("Left click release")
    elif event==2:
        print("Right click press")
    elif event==5:
        print("Right click release")
    elif event==8:
        print("Right double click press")

```

Event values:
mouse hover              >> 0

left click press         >> 1

left double click press  >> 7

left click release       >> 4

Right click press        >> 2

Right double click press >> 8

Right click release      >> 5

```python
cv2.imshow("window",image)
cv2.setMouseCallback("window",draw)
cv2.waitKey(0)
```
