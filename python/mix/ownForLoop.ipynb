Own For Loop

```python
def own_for_loop(iterable):
    iterator = iter(iterable)
    while True:
        try:
            print(next(iterator))
        except StopIteration:
            break
```

```python
own_for_loop([1,2,3,4,5])
```

How Iterator plays an important role in Memory saving:

```python
import sys

a= [n*2 for n in range(100000)]
```

```python

sys.getsizeof(a)
```

```python
for i in range(100000):
    a = i*2
    
sys.getsizeof(a)
```
